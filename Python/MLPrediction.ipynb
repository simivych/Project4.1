```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
data_exp = pd.read_csv('complete data - 10 games.csv')
data = pd.read_csv('data.csv')
data_exp
```

```python
data_exp['BestAI'] = data_exp[['numWins_Alpha-Beta Search','numWins_UCT', 'numWins_PN-MCTS', 'numWins_MAST', 'numWins_GRAVE']].idxmax(axis=1)
data_exp = data_exp[['Name', 'BestAI']]
X = pd.merge(data, data_exp, left_on='Game', right_on='Name')
X = X.drop(['Unnamed: 0','RulesetId', 'Game','MainRuleset','LudiiRuleset','GameId','Name_x','Name_y'], axis=1)
X
```

```python
import numpy as np
X['BestAI'].value_counts()
```

```python
agents = {
    'UCT': 0,
    'PN-MCTS': 1,
    'MAST': 2,
    'GRAVE': 3,
    'Alpha-Beta': 4
}
```

```python
y = []
for bestai in X.BestAI:
  for agent in agents:
    if agent in bestai:
      y.append(agents[agent])
      break
names = X[' Name']
X.drop(['BestAI',' Name'], axis=1, inplace=True)
```

```python
X = X.fillna(-1)
X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X, y)
```

```python
X_names_test = data['Game'] 
X_test = data.drop([' Name','Name','Unnamed: 0','RulesetId', 'Game','MainRuleset','LudiiRuleset','GameId','Name'], axis=1)
X_test = X_test.fillna(-1)
X_test
```

```python
y_pred = model.predict(X_test)
```

```python
predictions = pd.DataFrame({'Name':X_names_test, 'BestAI_pred':y_pred})
predictions
```

```python
predictions.to_csv('predictions.csv')
```
